```python
VIDEO_INPUT_PATH = Path("./ict3104-team11-2023/data/input_vid/")
VIDEO_RESIZE_PATH = Path("./ict3104-team11-2023/data/resize_vid/")
VIDEO_SKELE_PATH = Path("./ict3104-team11-2023/data/skele_vid/")

selected_video_path = None
resize_video_path = None
skele_video_path = None
superimpose_video_path = None
video_duration = 0

def get_subfolders(path):
    return [folder.name for folder in path.iterdir() if folder.is_dir()]

def get_mp4_files(path):
    return [file.name for file in path.iterdir() if file.suffix == '.mp4']

def play_processed_video(output_path, width=512, height=512):
    video_display = Video(str(output_path), width=width, height=height, embed=True)
    display(video_display)

def display_green_text(msg):
    display(HTML(f"<span style='color: green'>{msg}</span>"))

def show_files_from_path(path):
    global selected_video_path, video_duration
    video_files = get_mp4_files(path)

    if not video_files:
        print("No .mp4 files found.")
        return

    @interact
    def select_video(file=video_files):
        global selected_video_path, video_duration
        selected_video_path = path / file
        video = VideoFileClip(str(selected_video_path))
        video_duration = video.duration
        display(HTML("<span style='color: green'>This video file will be used for the next step.</span>"))
        play_processed_video(selected_video_path)

if VIDEO_INPUT_PATH.exists():
    subfolders = get_subfolders(VIDEO_INPUT_PATH)

    if not subfolders:
        print("No subfolders found in the main directory.")
    else:
        @interact
        def select_subfolder(subfolder=subfolders):
            show_files_from_path(VIDEO_INPUT_PATH / subfolder)
else:
    print("Main directory doesn't exist.")
```

### Video Converter
In this section, we will resize, clip and convert the video to be ready for MMPose. Do ensure that you have selected a video before running this cell. If you wish to change the selected video, do remember to re-run this cell.

We recommend a **max video length** of **8 seconds** for the best experience.

```python
start_time_input = FloatText(value=0.0, description='Start time:')
end_time_input = FloatText(description='End time:')

def resize_and_clip_video(video_path, output_folder, start_time, end_time):
    clip = VideoFileClip(str(video_path)).subclip(start_time, end_time)
    clip_resized = clip.resize((512, 512))
    output_path = Path(output_folder) / f"{Path(video_path).stem}_resize.mp4"

    clip_resized.write_videofile(str(output_path), verbose=False)

    clear_output(wait=True)
    display_green_text(f"Video processed and saved to {output_path}")
    global resize_video_path
    resize_video_path = output_path
    play_processed_video(output_path)


def on_button_click(button):
    if selected_video_path:
        clear_output(wait=True)
        display_green_text(f"Processing video {selected_video_path.name}...")

        resize_and_clip_video(selected_video_path, VIDEO_RESIZE_PATH, start_time_input.value, end_time_input.value)

process_button = widgets.Button(description="Process Video")
end_time_input.value = video_duration
process_button.on_click(on_button_click)
print("Video selected: " + str(selected_video_path))
input_box = HBox([start_time_input, end_time_input])
VBox([input_box, process_button])
```

```python
def generate_skeleton_video(video_path, output_folder):

    if ".mp4" not in str(video_path):
        print("Please choose a file with .mp4 format.")
        return

    file_name = video_path.stem
    skeleton_file_name = f"{file_name}_skeleton.mp4"

    skeleton_output_path = output_folder / skeleton_file_name

    inference_path = "./mmpose/demo/inferencer_demo.py"
    shell_command = f"{sys.executable} {inference_path} {video_path} --pose2d human --black-background --thickness 5 --vis-out-dir {skeleton_output_path}"

    proc = subprocess.Popen(shell_command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    i = 0
    for line in iter(proc.stdout.readline, ''):
        print(f"INFO:[{i}] " + line.strip())
        i+= 1
    # for _ in iter(proc.stdout.readline, ''):
    #     print("|", end='', flush=True)

    for line in iter(proc.stderr.readline, ''):
        print("ERROR: " + line.strip())

    # clear_output(wait=True) commented out for debugging purposes
    display_green_text(f"Skeleton video processed and saved to {skeleton_output_path}")
    global skele_video_path
    skele_video_path = skeleton_output_path
    play_processed_video(skele_video_path)

def on_generate_skeleton_click(button):
    clear_output(wait=True)
    display_green_text(f"Preparing {resize_video_path.name} for MMPose...")
    generate_skeleton_video(resize_video_path, VIDEO_SKELE_PATH)

generate_skeleton_button = Button(description="Generate Skeleton")
generate_skeleton_button.on_click(on_generate_skeleton_click)
display(generate_skeleton_button)
```

## Generating Dataset for Training Model 

This section sets up the necessary directory structure as specified in hdvila.py under(/part_{part_id}/video_clips). It involves transferring the MMPose-generated skeleton videos, which are based on the Charades dataset, into their respective folders.

Furthermore, a CSV file named caption_rm2048_train.csv is created and filled with details such as the storage paths for each video clip, the names of the clips, and captions provided by the user. This CSV will serve as a dataset for training the model.







```python
# Paths for the source and destination directories
source_dir = Path('/content/ict3104-team11-2023/data/skele_vid')
destination_dir = Path('/content/FollowYourPose/followyourpose/data')
# Iterate through the MMPose skeleton videos source path and count the number of videos
video_files = [f for f in os.listdir(source_dir) if f.endswith('.mp4')]
num_videos = len(video_files)

# Based on the number of MMPose skeleton videos, create the 'part_/video_clips' folders (this is based on the required file directory preset in hdvila.py)
for i, video_file in enumerate(video_files[:10]):
    part_id = i + 1
    video_clips_folder = os.path.join(destination_dir, f'part_{part_id}', 'video_clips')
    os.makedirs(video_clips_folder, exist_ok=True)

    source_path = os.path.join(source_dir, video_file)
    video_count = 1

# Copy MMPose skeleton videos into designated folders.
# Each video is stored in its respective 'part_X/video_clips' directory as per hdvila.py's requirements.
# The naming convention for the videos is sequential: '0001.mp4' for the first video, '0002.mp4' for the second, and so on.
# Since only one video corresponds to each 'part_X' directory, the video filename will always be '0001.mp4'.
    while True:
        destination_path = os.path.join(video_clips_folder, f'{video_count:04}.mp4')
        if not os.path.exists(destination_path):
            break
        video_count += 1

    shutil.copy(source_path, destination_path)

# Generate 'caption_rm2048_train.csv' to map video clips to their captions.
# CSV file includes 'part_id' and 'clip_id' to specify the video clip's location, and 'caption' to describe the content or context of the video clip.
# The 'part_id' correlates with the directory, while 'clip_id' corresponds to the specific video file within that directory.
csv_filename = 'caption_rm2048_train.csv'
csv_path = os.path.join(destination_dir, csv_filename)

with open(csv_path, 'w', newline='') as csv_file:
    fieldnames = ['part_id', 'clip_id', 'caption']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader()

    # Traverse each 'part_/video_clips' directory and record the details of video clips into 'caption_rm2048_train.csv'.
    # Every row in the CSV represents a video clip, with 'part_id', 'clip_id', and 'caption' as column data.
    for part_id in range(1, num_videos + 1):
        part_folder = os.path.join(destination_dir, f'part_{part_id}')
        video_clips_folder = os.path.join(part_folder, 'video_clips')

        # User to key in their desired caption for each video clip
        for clip_id, video_file in enumerate(os.listdir(video_clips_folder), start=1):
            video_path = os.path.join(video_clips_folder, f'{clip_id:04}.mp4')
            clip_id = f'{clip_id:04}'
            print(f"Please enter a new caption for {video_path} (press Enter after keying in):\n")
            caption = input()
            print("\n")

            writer.writerow({'part_id': part_id, 'clip_id': clip_id, 'caption': caption})

print(f"\n{csv_filename} has been created at: {csv_path}")
```

## Execution of Training 

This section executes the training command and saves all the newly trained model under content/ict3104-team11-2023/data/models/trained_model.

```python
# Set the file paths for the source videos and destination model using pathlib for improved path handling
csv_source_path = Path('/content/FollowYourPose/followyourpose/data/caption_rm2048_train.csv')
video_source_path = Path("/content/FollowYourPose/followyourpose/data")
model_destination_path = Path('/content/ict3104-team11-2023/data/models/trained_model')

# Read the caption CSV and store the data for later use
csv_data = []
with csv_source_path.open('r', newline='') as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        csv_data.append(row)

# Mark the start time for the training proces
start_time = time.time()

# Determine the number of videos to be processed by counting the files in each 'part_' directory
total_videos = sum(len(os.listdir(os.path.join(video_source_path, part_dir, "video_clips"))) for part_dir in os.listdir(video_source_path) if part_dir.startswith("part_"))

# Request a model name from the user before starting the process
input_model_name = input("\nName the new model (press Enter after naming):\n")

# Counter to keep track of the number of processed videos
processed_videos = 0

# Function to extract the numerical part_id from the directory name
def get_part_id(directory_name):
    return directory_name.replace("part_", "")

# Define a function to manage the video processing, configuration file setup, and model training
def process_video(video_info, video_path, part_id, clip_id_with_zeros, model_destination_path):
    print(f"Video info in video source path: Part ID: {part_id}, Clip ID: {clip_id_with_zeros}")
    print(f"Found matching video info in 'caption_rm2048_train.csv': {video_info}")
    print(f"Processing video for: {video_path}")

    # Load the pose_train.yaml
    with open('/content/FollowYourPose/configs/pose_train.yaml', 'r') as config_file:
        config = yaml.safe_load(config_file)

    # Update the skeleton_path in the config.yaml
    config['skeleton_path'] = f"./followyourpose/data/part_{part_id}/video_clips/{clip_id_with_zeros}.mp4"

    # Save the updated config.yaml in the new model
    # This will allow each iteration of the training command to replace the skeleton_path, before running it (to allow us to train different models across all generated skeletons)
    config_file_path = os.path.join(model_destination_path, f"config_{clip_id_with_zeros}_{part_id}.yaml")
    with open(config_file_path, 'w') as new_config_file:
        yaml.dump(config, new_config_file, default_flow_style=False)

    # Execute the training command on the video
    command = f"TORCH_DISTRIBUTED_DEBUG=DETAIL accelerate launch train_followyourpose.py --config='{config_file_path}' --skeleton_path='{video_path}'"
    result = subprocess.run(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    print(result.stdout.decode())
    print(result.stderr.decode())
    print(f"Finished processing video: {video_path}")
    print("\n")

    # Delete the config.yaml generated previously (to allow us to replace the originally generated config.yaml in the new model with the updated config.yaml)
    # This is used as a base config.yaml to generate the new config.yaml in each new model
    os.remove(config_file_path)

    # Copy the new model to the destination folder
    source_output_folder = model_destination_path

    # User input of desired model name
    input_model_name = input("\nEnter the new model name (press Enter after keying in):\n")

    # Save new model to the destination path with the new model named by user
    destination_output_folder = os.path.join(model_destination_path, f"{input_model_name}")
    shutil.copytree(source_output_folder, destination_output_folder)

    if result.returncode == 0:
        print(f"\nFinished processing video for: {video_path}")
        print("\n")
    else:
        print(f"Error processing video for:{video_path}")
        print("\n")

# Initialize a progress bar to visualize the training progress
progress_bar = tqdm(total=total_videos, desc='Processing Videos for Training')

# Set up a widget-based progress bar for visualization
progress_bar = widgets.FloatProgress(
    value=0.0,
    min=0.0,
    max=1.0,
    description='Progress for Training:',
    bar_style='info',
)

# Create a display widget to show the progress percentage
percentage_num = widgets.HTML(value="0.00%")

# Organize the progress bar and percentage display into a single horizontal box layout
progress_container = HBox([progress_bar, percentage_num])

# Display the progress container
display(progress_container)

# Iterate over each 'part_' directory to process and train a model for each corresponding video clip
# Update the progress bar accordingly after each video is processed
for part_dir in os.listdir(video_source_path):
    if part_dir.startswith("part_"):
        part_path = os.path.join(video_source_path, part_dir, "video_clips")
        part_id = get_part_id(part_dir)

        if os.path.exists(part_path):
            clip_counter = 0

            for clip_file in os.listdir(part_path):
                # Use the 'clip_id' from CSV data directly
                clip_id = clip_file.split('.')[0]

                # Check if the clip_id (file name) exists in the CSV data
                for row in csv_data:
                    if row['clip_id'] == clip_id and row['part_id'] == part_id:
                        video_info = row
                        # Construct the video path based on the clip_file and the base path
                        video_path = os.path.join(part_path, clip_file)

                        if os.path.exists(video_path):
                            clip_counter += 1
                            clip_id_with_zeros = str(clip_counter).zfill(4)

                            # Call the process_video function to handle video processing
                            process_video(video_info, video_path, part_id, clip_id_with_zeros, model_destination_path)

                            # Update the progress bar
                            processed_videos += 1
                            progress_percentage = (processed_videos / total_videos) * 100
                            progress_bar.value = processed_videos / total_videos

                            # Update the percentage label
                            percentage_num.value = f'{progress_percentage:.2f}%'

# # After processing all videos, remove the 'output' directory that served as a temporary model storage
# if processed_videos == total_videos:
#     output_folder_path = model_destination_path
#     shutil.rmtree(output_folder_path)

# Capture the end time to calculate the total runtime for training
end_time = time.time()
total_runtime = end_time - start_time

# Display the total runtime and a completion message
print(f"\nTotal Runtime: {total_runtime} seconds")
display(HTML('<h2>Processing completed!</h2>'))
```
